Given: A positive integer n≤105
 and a sorted array A[1..n]
 of integers from −105
 to 105
, a positive integer m≤105
 and a sorted array B[1..m]
 of integers from −105
 to 105
.

Return: A sorted array C[1..n+m]
 containing all the elements of A
 and  B

```python
with open("rosalind_mer.txt", "r") as f:
    n = int(f.readline())
    A = [int(i) for i in f.readline().strip().split(' ')]
    m = int(f.readline())
    B = [int(i) for i in f.readline().strip().split(' ')]
```

```python
def merge_sort(A, B):
    C = []
    i, j = 0, 0
    while i < len(A) and j < len(B):
        if A[i] <= B[j]:
            C.append(A[i])
            i += 1
        else:
            C.append(B[j])
            j += 1
	
    C += A[i:]
    C += B[j:]
    return C
C = merge_sort(A, B)
for i in C:
        print(i, end=" ")


```
